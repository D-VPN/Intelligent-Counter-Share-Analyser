```python
import cv2
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
shelf_1_column_1="column_0.png"
shelf_1_column_2="column_1.png"
shelf_2_column_1="shelf2/columns/column_0.png"
shelf_2_column_2="shelf2/columns/column_1.png"
shelf_2_column_3="shelf2/columns/column_2.png"
```

```python
img = cv2.imread(shelf_1_column_1)
```

## Converting image to grey scale and then applying threshold

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
img = cv2.threshold(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY),120,255,cv2.THRESH_TOZERO_INV)[1]
plt.imshow(img)
```

```python
from skimage import feature
```

```python
edges = feature.canny(
    image=img,
    sigma = 10,
    low_threshold=0.2,
    high_threshold=0.4
)
```

```python
plt.imshow(edges)
```
