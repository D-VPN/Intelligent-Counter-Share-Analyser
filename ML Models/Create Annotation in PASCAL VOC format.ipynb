```python
def GenerateXML(filename,count,x,y,w,h,category):
    path = annotation_ath;
    annotation = xm.Element("annotation")
    folder = xm.Element("folder")
    annotation.append(folder)
    folder.text = "images"
    fn = xm.Element("filename")
    annotation.append(fn)
    fn.text = filename +".JPG"
    source = xm.Element("source")
    annotation.append(source)
    db = xm.SubElement(source,"database")
    db.text ="Unknown"
    size = xm.Element("size")
    annotation.append(size)
    width = xm.SubElement(size,"width")
    width.text = "275"
    height = xm.SubElement(size,"height")
    height.text = "183"
    depth = xm.SubElement(size,"depth")
    depth.text = "3"
    segmented = xm.Element("segmented")
    annotation.append(segmented)
    segmented.text = "0"
    
    
    for i in range(count):
        obj = xm.Element('object')
        annotation.append(obj)
        name = xm.SubElement(obj,"name")
        name.text = str(category[i])
        
        pose = xm.SubElement(obj,"pose")
        pose.text = "Unspecified"
        
        truncated = xm.SubElement(obj,"truncated")
        truncated.text = "0"
        
        difficult = xm.SubElement(obj,"difficult")
        difficult.text = "0"
        
        bndbox = xm.SubElement(obj,"bndbox")
        xmin = xm.SubElement(bndbox,"xmin")
        xmin.text = str(x[i])
        ymin = xm.SubElement(bndbox,"ymin")
        ymin.text = str(y[i])
        xmax = xm.SubElement(bndbox,"xmax")
        xmax.text = str(x[i]+w[i])
        ymax = xm.SubElement(bndbox,"ymax")
        ymax.text = str(y[i]+h[i])
        
    tree = xm.ElementTree(annotation)
    with open(os.path.join(path,filename+'.xml'),"wb") as files:
        tree.write(files)
```

```python
for filename,group in df.groupby('filename'):
    x = []
    y = []
    w = []
    h = []
    category = []
    count=0
    for row_index,row in group.iterrows():
        count+=1
        x.append(row['x'])
        y.append(row['y'])
        w.append(row['w'])
        h.append(row['h'])
        category.append(row['category'])
    GenerateXML(filename[:-4],count,x,y,w,h,category)
```
